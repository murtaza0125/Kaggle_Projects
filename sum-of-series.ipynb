```python
def SeriesSum(x):
    listofterms = []
    term = 0
    S = 0
    if x==1:
        pass
    elif x>1:
        for n in range(2,(x+1)):
            term = (n/((math.factorial(x))*(x-(n-1))))
            listofterms.append(term)
    S = 1+ sum(listofterms)
    print(S)
    
```
