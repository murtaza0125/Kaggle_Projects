```python
data = pd.read_csv('../input/pima-indians-diabetes-database/diabetes.csv')
data.head()
```

```python
import seaborn as sns
```

```python
sns.heatmap(data.corr(), annot=True)
```

```python
X = data.iloc[:, 0:8]
y = data.iloc[:, 8]
```

```python
X.shape, y.shape
```

```python
# Split the dataset into training set and test set with 0.7 - 0.3 ratio

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
seed = 1
np.random.seed(seed)

#define the keras model

from keras.models import Sequential
from keras.layers import Dense

model = Sequential()

model.add(Dense(8, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='sgd', metrics='accuracy')
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, batch_size=10, epochs=300, validation_data=(X_test, y_test), verbose=1)
```

```python
history.history.keys()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize= (8,6))
plt.plot(history.epoch, history.history['loss'])
plt.plot(history.epoch, history.history['val_loss'])
plt.legend(['training_loss', 'val_loss'])
plt.ylim(0.5,1)
plt.xlabel('Epochs', fontsize=15)
plt.ylabel('Loss Function Val', fontsize=15)
plt.grid('on')
plt.show()
```

```python
plt.figure(figsize= (8,6))
plt.plot(history.epoch, history.history['accuracy'])
plt.plot(history.epoch, history.history['val_accuracy'])
plt.legend(['accuracy', 'val_accuracy'])

plt.xlabel('Epochs', fontsize=15)
plt.ylabel('Accuracy', fontsize=15)
plt.grid('on')
plt.show()
```

```python
seed = 1
np.random.seed(seed)

#define the keras model

from keras.models import Sequential
from keras.layers import Dense
from keras.regularizers import l2

model = Sequential()

model.add(Dense(8, input_dim=8, activation='relu', kernel_regularizer=l2(0.01)))
model.add(Dense(8, activation='relu', kernel_regularizer=l2(0.01)))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='sgd', metrics='accuracy')

history=model.fit(X_train, y_train, batch_size=10, epochs=300, validation_data=(X_test, y_test), verbose=0)
```

```python
plt.figure(figsize= (8,6))
plt.plot(history.epoch, history.history['loss'])
plt.plot(history.epoch, history.history['val_loss'])
plt.legend(['training_loss', 'val_loss'])
plt.ylim(0.5,1)
plt.xlabel('Epochs', fontsize=15)
plt.ylabel('Loss Function Val', fontsize=15)
plt.grid('on')
plt.show()
```

```python
plt.figure(figsize= (8,6))
plt.plot(history.epoch, history.history['accuracy'])
plt.plot(history.epoch, history.history['val_accuracy'])
plt.legend(['accuracy', 'val_accuracy'])

plt.xlabel('Epochs', fontsize=15)
plt.ylabel('Accuracy', fontsize=15)
plt.grid('on')
plt.show()
```
